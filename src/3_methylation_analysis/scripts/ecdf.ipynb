```python
import numpy as np
import matplotlib.pyplot as plt
# chr1    28182777        s31222:0.000    a509:0.000      a509_2:0.067
# 生成示例数据
data1 = np.loadtxt("/public/home/xiayini/project/nanopore_ATRX/3_methylation_analysis/results/s31222_a509_a509_2/three_locations/chr1.modifis_grow_sam1.txt")
data2 = np.loadtxt("/public/home/xiayini/project/nanopore_ATRX/3_methylation_analysis/results/s31222_a509_a509_2/three_locations/chr1.modifis_grow_sam2.txt")
data3 = np.loadtxt("/public/home/xiayini/project/nanopore_ATRX/3_methylation_analysis/results/s31222_a509_a509_2/three_locations/chr1.modifis_grow_sam3.txt")
print(data1)
# 对数据进行排序
sorted_data1 = np.sort(data1)
sorted_data2 = np.sort(data2)
sorted_data3 = np.sort(data3)

# 计算累积概率
ecdf_values1 = np.arange(1, len(sorted_data1) + 1) / len(sorted_data1)
ecdf_values2 = np.arange(1, len(sorted_data2) + 1) / len(sorted_data2)
ecdf_values3 = np.arange(1, len(sorted_data3) + 1) / len(sorted_data3)

# 绘制经验累积函数图
plt.plot(sorted_data1, ecdf_values1, marker='.', linestyle='none', label='Data 1')
plt.plot(sorted_data2, ecdf_values2, marker='.', linestyle='none', label='Data 2')
plt.plot(sorted_data3, ecdf_values3, marker='.', linestyle='none', label='Data 3')
plt.xlabel('Data')
plt.ylabel('ECDF')
plt.legend()
plt.grid(True)
plt.show()
```
